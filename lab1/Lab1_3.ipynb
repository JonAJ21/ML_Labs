3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def norm(matr):
    A = np.array(matr)
    sum = np.sum(A) # \sum_{i} \sum_{j} A_{ij}
    if sum == 0:
        raise ValueError("Sum of all elements is zero")
    return A / sum


# Example
matrix = [[1, 2, 3], [4, 5, 6]]
norm(matrix)
```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def piramid_volume(X, Y, Z): 
    A = np.array([X, Y, Z])
    detA = np.linalg.det(A)
    return np.abs(detA) / 6

# Example

X = np.array([3, 1, 2])
Y = np.array([6, 3, 2])
Z = np.array([3, -2, 2])

piramid_volume(X, Y, Z)
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def boundary_matr(N):
    A = np.zeros((N, N), dtype = int)
    A[0, :] = 1
    A[-1, :] = 1
    A[:, 0] = 1
    A[:, -1] = 1
    return A

print(boundary_matr(1), "\n")
print(boundary_matr(2), "\n")
print(boundary_matr(3), "\n")
print(boundary_matr(4), "\n")
print(boundary_matr(5), "\n")
```

7. Как найти наиболее близкое значение (к заданному скаляру) в векторе?

```python
def closest_val(vector, scalar):
    idx = np.argmin(np.abs(vector - scalar))
    return vector[idx]

vec = np.random.randint(1, 100, size = 10) # Создаем рандомный вектор
print(vec)

print(closest_val(vec, 41))
print(closest_val(vec, 52))
print(closest_val(vec, 63))
print(closest_val(vec, 100))


```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
diseases_by_continent = filled_merged_data.groupby("continent")["total_cases"].sum()

plt.figure(figsize=(5, 5))
plt.pie(diseases_by_continent, labels=diseases_by_continent.index, autopct="%1.1f%%")
plt.title("Число заболеваний COVID-19 по континентам")
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
df = covid_data[covid_data["location"].isin([
    "Russia",
    "China",
    "United States",
    "India",
    "Nigeria"
])].sort_values(by="total_tests", ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(df["location"], df["total_tests"], color="sandybrown")
plt.xlabel("County")
plt.ylabel("Total tests")
plt.title("Total COVID-19 tests in Russia, China, United States, India, Nigeria")
plt.show()

print(f"Countries without information about total tests:\n{df.loc[df.isna().any(axis=1)]["location"]}")

```

6. Сравните распределения кол-ва смертей на 1000 человек в Европе и Африке

```python
filled_merged_data["mortality_per_thousand"] = filled_merged_data["mortality_per_million"] / 1000

europe_df = filled_merged_data[filled_merged_data["continent"] == "Europe"]
africa_df = filled_merged_data[filled_merged_data["continent"] == "Africa"]

plt.figure(figsize=(10, 5))
sns.kdeplot(europe_df["mortality_per_thousand"].dropna(), fill=True, label="Europe")
sns.kdeplot(africa_df["mortality_per_thousand"].dropna(), fill=True, label="Africa")

plt.title("Distribution of deaths per 1000 people in Europe and Africa")
plt.legend(loc="upper right")
plt.show()

```
